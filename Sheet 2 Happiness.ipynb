```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read sheets 2 and 3 of the Excel file into separate DataFrames
df1 = pd.read_excel('edited.xlsx', sheet_name=0)
df2 = pd.read_excel('edited.xlsx', sheet_name=1)
df3 = pd.read_excel('edited.xlsx', sheet_name=2)

# Print the first few rows of each DataFrame
print('Sheet 1:')
print(df1.head())
print('Sheet 2:')
print(df2.head())
print('Sheet 3:')
print(df3.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the table to get 'Scored by' as index and 'Week' as columns
df_pivot = df2.pivot(index='Parameters', columns='week')

# Create heatmap using seaborn
plt.figure(figsize=(32,10))
sns.heatmap(df_pivot, cmap='coolwarm',annot=True,annot_kws={"size": 25})

# Add title and axis labels
plt.title('Happiness Scores',fontsize=25)
plt.xlabel('Employee-Week',fontsize=25)
plt.ylabel('Parameters',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=10)


#plt.savefig('Heatmap Scores.png')
# Show plot
plt.savefig('happiness weekly scores heatmap.jpg')

plt.show()
```

```python
df2.set_index('Parameters', inplace=True)
```

```python
groups = df2.groupby('week')
# plot each group separately
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,8), sharey=True)
for i, (week, group) in enumerate(groups):
    group = group.drop('week', axis=1)
    ax = axs[i]
    group.plot(kind='bar', ax=ax)
    ax.set_xlabel("Parameters")
    ax.set_ylabel("Score")
    ax.set_title(f"Week {week}")
plt.legend(bbox_to_anchor=(1.0, 0.5))

    
# show the plot
plt.tight_layout()
plt.savefig('Happiness weekly score.jpg' )
plt.show()
```
