```python
import matplotlib.pyplot as plt
import numpy as np

# data for the chart
parameters = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5','Q6', 'Q7', 'Q8', 'Q9', 'Q10']
neha_scores = [10, 9, 9, 9, 9,9,9,9,9,9]
amit_scores = [8, 7, 8, 8, 8,8,8,8,8,8]
sachin_scores = [6, 4, 7, 6, 6,7,7,6,6,6]

plt.figure(figsize=(11,6))
# set the width of the bars
bar_width = 0.25

# set the positions of the x-axis ticks (center of the bars)
r1 = np.arange(len(parameters))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# create the bar chart
plt.bar(r1, neha_scores, color='yellow', width=bar_width, edgecolor='black', label='Neha')
plt.bar(r2, amit_scores, color='red', width=bar_width, edgecolor='black', label='Amit')
plt.bar(r3, sachin_scores, color='blue', width=bar_width, edgecolor='black', label='Sachin')

# add labels, title, and legend
plt.xlabel('Parameters')
plt.ylabel('Scores')
plt.xticks([r + bar_width for r in range(len(parameters))], parameters)
plt.title('Progress Week 1 Scores')
plt.legend(bbox_to_anchor=(1.0, 0.5))

# show the chart
plt.savefig('Progress week1 scores.jpg')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart
categories = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5','Q6', 'Q7', 'Q8', 'Q9', 'Q10']
neha_scores = [10, 9, 9, 9, 9,9,9,9,9,9]
amit_scores = [8, 7, 8, 8, 8,8,8,8,8,8]
sachin_scores = [6, 4, 7, 6, 6,7,7,6,6,6]

# Number of categories
N = len(categories)

# Calculate angles for each category
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

# Initialise the radar chart
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

# Add the first scores to the chart
values = neha_scores + neha_scores[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Neha")
ax.fill(angles, values, 'b', alpha=0.1)

# Add the second scores to the chart
values = amit_scores + amit_scores[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Amit")
ax.fill(angles, values, 'r', alpha=0.1)

# Add the third scores to the chart
values = sachin_scores + sachin_scores[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Sachin")
ax.fill(angles, values, 'g', alpha=0.1)

# Add the legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Set the title of the chart
plt.title("Week 1 Radar Chart")

# Show the chart
plt.savefig('Progress week1 Radar Chart.jpg')
plt.show()
```
